```python
#remoe uppercse words, stop words, and words with length less than 5 
from nltk.corpus import stopwords
import re 
nltk.download('stopwords')
upper = re.findall(r'\b[A-Z]+(?:\s+[A-Z]+)*\b', content)
#print(upper)
stopword = stopwords.words('english')
new_toke = [word for word in tokens if word not in upper]
new_2 = [word for word in new_toke if word not in stopword]
new_3 = [word for word in new_2 if len(word) > 5]
tokens= [word for word in new_3 if word.isalpha()]
print (tokens)
```
